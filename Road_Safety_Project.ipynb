## Abstract
This project investigates the relationship between traffic intensity, road infrastructure, and road accidents in Bulgaria. A data-driven approach is applied using statistical analysis and hypothesis testing to evaluate whether increased traffic leads to higher accident rates and whether infrastructure improvements influence accident severity.

---

## Keywords
Road Safety, Traffic Analysis, Accident Rates, Infrastructure, Data Analysis, Hypothesis Testing

## 1. Introduction

Road safety is a critical issue in modern transportation systems. The increase in vehicle numbers and changes in infrastructure may influence accident rates.

This study aims to analyze the relationship between:
- Traffic intensity
- Road infrastructure
- Road accidents

The focus is on Bulgaria, using real-world datasets.

## 2. Problem Formulation

The main research questions are:

- Does increased traffic lead to more road accidents?
- Does improved infrastructure reduce accident severity?

### Assumptions
- Data is accurate and representative
- External factors (weather, behavior) are not included

### Limitations
- Limited dataset granularity
- Possible inconsistencies in reporting

## 3. Data Sources

Two independent datasets are used:

### Dataset 1: Road Accidents
- Source: National Statistical Institute / Ministry of Interior
- Variables: accidents, fatalities, injured

### Dataset 2: Traffic and Infrastructure
- Source: Road Infrastructure Agency / Eurostat
- Variables: vehicles, road length
