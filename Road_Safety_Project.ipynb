## 2. Problem Formulation

Road safety is a complex and multifactorial problem influenced by traffic intensity, infrastructure characteristics, human behavior, and external conditions. Despite significant efforts to improve transportation systems, road accidents remain a critical societal and engineering challenge.

The present study focuses on the relationship between traffic intensity, road infrastructure, and road accident dynamics in Bulgaria. The core objective is to investigate whether observable statistical relationships exist between these factors and to what extent they can be quantified using available data.

More specifically, the analysis addresses the following research questions:

- Is there a statistically significant relationship between traffic intensity and the number of road accidents?
- How do normalized indicators (e.g., accidents per vehicle or per unit road length) evolve over time?
- Can changes in infrastructure be associated with variations in accident frequency or severity?

To answer these questions, the real-world problem is translated into a quantitative framework where traffic, infrastructure, and accident data are represented as measurable variables. Statistical tools are then applied to evaluate relationships between these variables.

It is important to note that the scope of this study is limited to identifying statistical associations rather than establishing causal relationships. The analysis is designed to provide insight into observable patterns in the data while maintaining a careful and disciplined interpretation of the results.

---

## 3. Assumptions

The analysis is based on the following assumptions:

1. **Data Reliability**  
   The datasets used are assumed to be accurate and representative of the underlying real-world phenomena within the limits of official reporting practices.

2. **Consistency Over Time**  
   It is assumed that data collection methodologies remain sufficiently consistent across the analyzed time period, allowing for meaningful temporal comparisons.

3. **Independence of Data Sources**  
   The two datasets (accidents and traffic/infrastructure) are treated as independent sources, each capturing different aspects of the system under study.

4. **Aggregation Validity**  
   Annual aggregated data is assumed to be adequate for identifying general trends, despite the loss of finer temporal or spatial resolution.

5. **Uniform Exposure**  
   Traffic indicators (e.g., number of vehicles) are used as proxies for exposure to risk, assuming that higher traffic levels correspond to higher potential interaction between road users.

6. **Neglected External Factors**  
   Factors such as weather conditions, driver behavior, enforcement intensity, and technological changes are not explicitly included in the model and are assumed to have an indirect or unobserved influence.

7. **Stationarity of Relationships**  
   The relationships between variables are assumed to be relatively stable over the observed period, allowing for the application of standard statistical methods such as correlation analysis.

---

## 4. Limitations

Despite the structured approach, several limitations must be acknowledged.

First, the analysis relies on aggregated annual data, which limits the level of detail and may obscure regional variations, short-term dynamics, and localized risk factors.

Second, potential inconsistencies in data collection, reporting standards, or measurement practices across different institutions may introduce uncertainty into the results.

Third, the exclusion of important explanatory variables — such as weather conditions, driver behavior, enforcement policies, and vehicle technology — limits the ability to fully explain observed patterns.

As a result, the findings of this study should be interpreted as indicative rather than definitive, and no causal conclusions should be drawn solely based on the presented analysis.

## 6. Methodology

The present study follows a structured data-driven methodology aimed at investigating the relationship between traffic intensity, road infrastructure, and road safety indicators in Bulgaria.
The methodology prioritizes normalized indicators in order to reduce bias caused by differences in population size, traffic volume, and infrastructure scale.

### Data Acquisition and Integration
Two independent datasets are utilized in this analysis: one describing road accident statistics and another containing traffic and infrastructure-related indicators. The datasets are collected from official sources and are assumed to be reliable within the limits of their reporting methodologies.

To enable joint analysis, the datasets are harmonized and merged based on common temporal identifiers (e.g., year). This integration step ensures consistency while preserving the independence of the original data sources.

### Data Preprocessing
Prior to analysis, the data undergoes systematic preprocessing, including:
- standardization of variable names and formats,
- handling of missing or inconsistent values,
- validation of numerical ranges and logical constraints.

Basic validation tests are applied to ensure data integrity and to prevent propagation of errors into subsequent stages of the analysis.

### Feature Engineering
To enable meaningful comparisons, several derived indicators are constructed. These include normalized metrics such as:
- accidents per unit of traffic (e.g., per 10,000 vehicles),
- accident severity indicators (e.g., fatalities per accident),
- infrastructure efficiency measures (e.g., accidents per kilometer of road network).

These transformations allow for the analysis of relative risk rather than relying solely on absolute values.

### Exploratory Data Analysis
Exploratory techniques are applied to identify trends, patterns, and potential anomalies in the data. Visualization and summary statistics are used to gain an initial understanding of the relationships between variables and to guide further analysis.

### Statistical Analysis
The primary analytical approach is based on correlation analysis, specifically the Pearson correlation coefficient, used to quantify the strength and direction of relationships between key variables such as traffic volume and accident frequency.

In addition, hypothesis testing is performed to evaluate the statistical significance of observed relationships. A null hypothesis assuming no association between variables is tested against an alternative hypothesis indicating the presence of a relationship.

Where appropriate, confidence intervals are estimated to quantify uncertainty and provide a range of plausible values for key metrics.

### Analytical Rigor and Limitations
The methodology emphasizes careful interpretation of results and avoids overgeneralization. The analysis is limited to observable statistical relationships and does not attempt to establish causal mechanisms.

All conclusions are drawn with consideration of:
- data aggregation levels,
- potential measurement errors,
- omitted variables and external factors not captured in the datasets.

This approach ensures a disciplined analytical framework and maintains a high level of epistemic rigor throughout the study.

```python
# =========================
# Bulgaria - fatalities dataset
# =========================

fatalities_bg = Road_fatalities_BG[["Year", "fatalities"]].copy()
fatalities_bg = fatalities_bg.rename(columns={"Year": "year"})


# =========================
# Population (exposure proxy)
# =========================

population_bg = Population_bg_2009_2024[["Year", "Total"]].copy()
population_bg = population_bg.rename(columns={
    "Year": "year",
    "Total": "population"
})


# =========================
# Merge datasets
# =========================

df_bg = fatalities_bg.merge(population_bg, on="year", how="left")

df_bg["road_km"] = 19986


# =========================
# Derived metric
# =========================

df_bg["fatalities_per_million"] = (
    df_bg["fatalities"] / df_bg["population"] * 1_000_000
)


# =========================
# Clean missing values (IMPORTANT)
# =========================

df_bg = df_bg.dropna(subset=["fatalities_per_million"])

df_bg.head()
```
