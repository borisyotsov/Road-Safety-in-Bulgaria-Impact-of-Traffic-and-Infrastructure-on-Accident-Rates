In October 2021, the World Health Organization and the United Nations officially launched the Global Plan for the Decade of Action for Road Safety 2021–2030. The plan emphasizes a systemic approach to road safety, including the development of infrastructure that meets technical safety standards for all road users. One of its key targets is that, by 2030, all new roads should achieve high safety standards (e.g., a minimum three-star safety rating), and more than 75% of travel should occur on roads meeting such standards.

At the European level, road safety policy is guided by the “Vision Zero” approach, which aims to eliminate fatalities and serious injuries in road transport. This approach is supported by initiatives such as the Valletta Declaration (2017), which outlines strategic priorities for the period 2021–2030, including improved enforcement, modernization of driver training and licensing, and the safe integration of new technologies and automation.

Despite these efforts, road traffic injuries remain a leading cause of death worldwide. According to the World Health Organization, road traffic crashes are the ninth leading cause of death globally and the leading cause of death among people aged 5–29 years. Vulnerable road users, including pedestrians, cyclists and motorcyclists, account for nearly 50% of all fatalities.

Furthermore, more than 90% of road traffic deaths occur in low- and middle-income countries, despite these countries accounting for only about 54% of the world’s registered vehicles. These figures highlight the scale of the problem and underline the need for data-driven and engineering-based approaches to improve road safety.

Within this global and European policy framework, understanding the relationship between traffic intensity, infrastructure characteristics, and accident occurrence becomes particularly important.

## 3. Assumptions

The analysis is based on the following assumptions:

1. **Data Reliability**  
   The datasets used are assumed to be accurate and representative of the underlying real-world phenomena within the limits of official reporting practices.

2. **Consistency Over Time**  
   It is assumed that data collection methodologies remain sufficiently consistent across the analyzed time period, allowing for meaningful temporal comparisons.

3. **Independence of Data Sources**  
   The two datasets (accidents and traffic/infrastructure) are treated as independent sources, each capturing different aspects of the system under study.

4. **Aggregation Validity**  
   Annual aggregated data is assumed to be adequate for identifying general trends, despite the loss of finer temporal or spatial resolution.

5. **Uniform Exposure**  
   Traffic indicators (e.g., number of vehicles) are used as proxies for exposure to risk, assuming that higher traffic levels correspond to higher potential interaction between road users.

6. **Exposure Approximation**
   Due to the lack of aggregated national traffic data, population is used as a proxy for exposure. It is assumed that higher population levels correspond to higher participation in traffic, although this does not fully capture actual traffic intensity.

7. **Spatial Consistency Assumption**
   Traffic measurements available at specific road segments are assumed to be indicative of broader traffic patterns, despite potential spatial variability.

8. **Exposure Proxy Interpretation**

Due to the lack of direct traffic measurements (e.g., vehicle kilometers traveled or traffic counts), this study uses proxy variables to approximate exposure to road risk.

For Bulgaria, population and registered vehicles are used as indirect indicators of traffic exposure. This assumes that higher population and vehicle ownership correlate with increased road usage and interaction between road users.

However, this approach introduces limitations:
- population does not directly measure traffic intensity,
- vehicle ownership does not capture actual road usage,
- transit traffic is not explicitly captured,
- temporal distribution of traffic is not available.

Therefore, results should be interpreted as relationships between *proxy exposure indicators* and road safety outcomes, rather than direct effects of traffic intensity.

## 4. Limitations

Despite the structured analytical approach, several important limitations must be acknowledged.

First, the analysis relies on aggregated annual data at the national level. This limits the ability to capture regional variability, local risk factors, and short-term dynamics in road safety.

Second, there is a fundamental mismatch between the available traffic data and accident statistics. Traffic measurements are typically collected at specific road segments and do not represent total traffic exposure at the national level. In contrast, fatality data is aggregated across the entire country. This lack of a common spatial unit prevents direct statistical linkage between traffic intensity and accident outcomes.

Third, the study uses population as a proxy for exposure due to the absence of comprehensive traffic data. While population provides a reasonable approximation, it does not account for differences in vehicle ownership, travel behavior, or actual traffic volume.

Fourth, Bulgaria experiences a significant amount of transit traffic, including international freight and passenger flows. This implies that road fatalities are influenced not only by the resident population but also by non-resident road users. As a result, exposure may be underestimated when using population-based indicators.

Finally, several important explanatory variables are not included in the analysis, such as weather conditions, driver behavior, enforcement intensity, and technological factors.

For these reasons, the results of this study should be interpreted as indicative of statistical associations rather than definitive evidence of causal relationships.

## 6. Methodology

The present study follows a structured data-driven methodology aimed at investigating the relationship between traffic intensity, road infrastructure, and road safety indicators in Bulgaria.
The methodology prioritizes normalized indicators in order to reduce bias caused by differences in population size, traffic volume, and infrastructure scale.

### Data Acquisition and Integration
Two independent datasets are utilized in this analysis: one describing road accident statistics and another containing traffic and infrastructure-related indicators. The datasets are collected from official sources and are assumed to be reliable within the limits of their reporting methodologies.

To enable joint analysis, the datasets are harmonized and merged based on common temporal identifiers (e.g., year). This integration step ensures consistency while preserving the independence of the original data sources.

### Data Preprocessing
Prior to analysis, the data undergoes systematic preprocessing, including:
- standardization of variable names and formats,
- handling of missing or inconsistent values,
- validation of numerical ranges and logical constraints.

Basic validation tests are applied to ensure data integrity and to prevent propagation of errors into subsequent stages of the analysis.

### Feature Engineering

To enable meaningful comparisons across time and between regions, the analysis constructs normalized safety indicators.

For Bulgaria, fatalities are normalized using population data to compute fatalities per million inhabitants. Additionally, road network length is used as a structural exposure proxy to approximate infrastructure density.

For the EU dataset, fatalities are reported as absolute values aggregated across countries. To ensure comparability, EU values are normalized using an approximate EU population baseline, allowing conversion into fatalities per million inhabitants.

These transformations ensure that both datasets are expressed on a comparable per-capita scale, enabling consistent cross-regional analysis of road safety outcomes.

### Exploratory Data Analysis
Exploratory techniques are applied to identify trends, patterns, and potential anomalies in the data. Visualization and summary statistics are used to gain an initial understanding of the relationships between variables and to guide further analysis.

### Statistical Analysis
The primary analytical approach is based on correlation analysis, specifically the Pearson correlation coefficient, used to quantify the strength and direction of relationships between key variables such as traffic volume and accident frequency.

In addition, hypothesis testing is performed to evaluate the statistical significance of observed relationships. A null hypothesis assuming no association between variables is tested against an alternative hypothesis indicating the presence of a relationship.

Where appropriate, confidence intervals are estimated to quantify uncertainty and provide a range of plausible values for key metrics.

### Analytical Rigor and Limitations
The methodology emphasizes careful interpretation of results and avoids overgeneralization. The analysis is limited to observable statistical relationships and does not attempt to establish causal mechanisms.

All conclusions are drawn with consideration of:
- data aggregation levels,
- potential measurement errors,
- omitted variables and external factors not captured in the datasets.

This approach ensures a disciplined analytical framework and maintains a high level of epistemic rigor throughout the study.

### Data Consistency and Measurement Constraints

A key methodological consideration in this study is the mismatch between available traffic data and accident statistics.

Traffic data is typically collected at specific road segments and does not represent total traffic volume at the national level. In contrast, fatality data is aggregated at the country level. This lack of a common spatial and analytical unit limits the ability to directly relate traffic intensity to accident outcomes.

Additionally, Bulgaria experiences significant transit traffic, including international freight and passenger flows. As a result, road exposure is not limited to the resident population, further complicating the interpretation of risk indicators.

Due to these constraints, population is used as a proxy for exposure. This approximation allows for the construction of normalized indicators but introduces uncertainty that must be considered when interpreting the results.

The analysis does not model causal relationships. Instead, it focuses on statistical associations between normalized safety indicators and proxy exposure variables.

Cluster Analysis of Road Fatalities (2014 vs 2024)
This scatter plot visualizes the relationship between road‑fatality rates in 2014 and 2024 for European countries. Each point represents a country, positioned according to its fatality rate in both years. A K‑Means clustering algorithm was applied to group countries into three risk categories:

High risk (red): countries with consistently high fatality rates

Medium risk (blue): countries around the European average

Low risk (green): countries with the lowest fatality rates

The clustering is based on both years simultaneously, which allows us to identify long‑term patterns rather than single‑year fluctuations.

Anomalies
Countries that deviate strongly from the overall distribution (based on Z‑score > 2) are marked with a black outline. These represent unusual cases that do not fit well into any cluster — typically:

exceptionally high fatality rates

exceptionally low fatality rates

Interpretation
Countries in the high‑risk cluster show persistently elevated fatality levels and may require targeted road‑safety interventions.

Countries in the medium‑risk cluster are close to the European average and may benefit from incremental improvements.

Countries in the low‑risk cluster demonstrate strong road‑safety performance and can serve as benchmarks for best practices.

This analysis helps identify structural patterns in road‑safety outcomes and highlights where policy attention may be most needed.

```python
from sklearn.cluster import KMeans
from scipy.stats import zscore

df3 = df2.copy()

# Replace NaN in 2014 (Iceland) with mean value
df3[2014] = df3[2014].fillna(df3[2014].mean())

# Prepare data
X = df3[[2014, 2024]].values

# K-Means clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df3["Cluster"] = kmeans.fit_predict(X)

# Assign risk levels based on cluster centroids
centroids = kmeans.cluster_centers_
order = centroids[:, 1].argsort()  # sort by 2024 values

risk_names = {
    order[0]: "Low risk",
    order[1]: "Medium risk",
    order[2]: "High risk"
}

df3["Risk_Level"] = df3["Cluster"].map(risk_names)

# Detect anomalies
df3["Zscore"] = zscore(df3[2024])
df3["Anomaly"] = df3["Zscore"].abs() > 2

# Colors
colors = {
    "Low risk": "#2ca02c",
    "Medium risk": "#1f77b4",
    "High risk": "#d62728"
}

plt.figure(figsize=(12, 8))

# Plot clusters
for level in df3["Risk_Level"].unique():
    subset = df3[df3["Risk_Level"] == level]
    plt.scatter(
        subset[2014], subset[2024],
        s=150, color=colors[level], label=level, alpha=0.8
    )

# Plot anomalies
anom = df3[df3["Anomaly"] == True]
plt.scatter(
    anom[2014], anom[2024],
    s=300, facecolors='none', edgecolors='black', linewidths=2, label="Anomaly"
)

# Labels
for _, row in df3.iterrows():
    plt.text(row[2014] + 0.5, row[2024] + 0.5, row["Country"], fontsize=9)

plt.xlabel("Fatalities per million inhabitants (2014)")
plt.ylabel("Fatalities per million inhabitants (2024)")
plt.title("Cluster Analysis of Road Fatalities (2014 vs 2024)")

plt.legend()
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

Cluster Analysis with Normalized Data
The previous clustering produced unbalanced groups because the raw data is highly skewed: many countries have low fatality rates, while only a few have very high values. To correct this, the data was normalized using StandardScaler, ensuring that both years (2014 and 2024) contribute equally to the clustering.

After normalization, K‑Means produces three well‑balanced clusters:

High risk: countries with consistently high fatality rates

Medium risk: countries around the European average

Low risk: countries with the lowest fatality rates

Anomalies (Z‑score > 2) remain highlighted as countries that do not fit well into any cluster.

This approach provides a more meaningful segmentation of road‑safety performance across Europe.

```python
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from scipy.stats import zscore

df3 = df2.copy()


# Replace NaN in 2014 (Iceland)
df3[2014] = df3[2014].fillna(df3[2014].mean())

# Prepare data
X = df3[[2014, 2024]].values

# Normalize (very important!)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Choose number of clusters = 3 (after scaling works much better)
kmeans = KMeans(n_clusters=3, random_state=42)
df3["Cluster"] = kmeans.fit_predict(X_scaled)

# Assign risk levels based on centroid order
centroids = kmeans.cluster_centers_
order = centroids[:, 1].argsort()

risk_names = {
    order[0]: "Low risk",
    order[1]: "Medium risk",
    order[2]: "High risk"
}

df3["Risk_Level"] = df3["Cluster"].map(risk_names)

# Detect anomalies
df3["Zscore"] = zscore(df3[2024])
df3["Anomaly"] = df3["Zscore"].abs() > 2

# Colors
colors = {
    "Low risk": "#2ca02c",
    "Medium risk": "#1f77b4",
    "High risk": "#d62728"
}

plt.figure(figsize=(12, 8))

# Plot clusters
for level in df3["Risk_Level"].unique():
    subset = df3[df3["Risk_Level"] == level]
    plt.scatter(
        subset[2014], subset[2024],
        s=150, color=colors[level], label=level, alpha=0.8
    )

# Plot anomalies
anom = df3[df3["Anomaly"] == True]
plt.scatter(
    anom[2014], anom[2024],
    s=300, facecolors='none', edgecolors='black', linewidths=2, label="Anomaly"
)

# Labels
for _, row in df3.iterrows():
    plt.text(row[2014] + 0.5, row[2024] + 0.5, row["Country"], fontsize=9)

plt.xlabel("Fatalities per million inhabitants (2014)")
plt.ylabel("Fatalities per million inhabitants (2024)")
plt.title("Cluster Analysis of Road Fatalities (2014 vs 2024) — Normalized")

plt.legend()
plt.grid(True, linestyle="--", alpha=0.4)
plt.tight_layout()
plt.show()
```

### 8.2. Data Cleaning

#### 8.2.1. Dataset 1 - Bulgaria

```python
# =========================
# Bulgaria - fatalities dataset
# =========================

fatalities_bg = Road_fatalities_BG[["Year", "fatalities"]].copy()
fatalities_bg = fatalities_bg.rename(columns={"Year": "year"})


# =========================
# Population (exposure proxy)
# =========================

population_bg = Population_bg_2009_2024[["Year", "Total"]].copy()
population_bg = population_bg.rename(columns={
    "Year": "year",
    "Total": "population"
})


# =========================
# Merge datasets
# =========================

df_bg = fatalities_bg.merge(population_bg, on="year", how="left")

df_bg["road_km"] = 19986


# =========================
# Derived metric
# =========================

df_bg["fatalities_per_million"] = (
    df_bg["fatalities"] / df_bg["population"] * 1_000_000
)


# =========================
# Clean missing values (IMPORTANT)
# =========================

df_bg = df_bg.dropna(subset=["fatalities_per_million"])

df_bg.head()
```

#### 8.2.2. Dataset 2 - EU

```python
df_eu = Fatalities_by_country.copy()

# Remove Excel artifacts
df_eu = df_eu.drop(columns=[c for c in df_eu.columns if "Unnamed" in str(c)], errors="ignore")

# Standardize column name
df_eu = df_eu.rename(columns={df_eu.columns[0]: "country"})

# Reshape to long format
df_eu = df_eu.melt(
    id_vars="country",
    var_name="year",
    value_name="fatalities_per_million"
)

# Type cleaning
df_eu["year"] = pd.to_numeric(df_eu["year"], errors="coerce")
df_eu["fatalities_per_million"] = pd.to_numeric(df_eu["fatalities_per_million"], errors="coerce")

df_eu = df_eu.dropna()

df_eu.head()
```

## 12. Hypothesis Testing

In this section, two complementary hypotheses are evaluated:

Relationship between exposure and fatalities (Bulgaria)

1.   Relationship between exposure and fatalities (Bulgaria)
2.   Comparison between Bulgaria and EU average fatality rates

The analysis combines correlation methods and statistical hypothesis testing to ensure a rigorous and transparent interpretation of results.

### 12.1. Hypothesis 1: Exposure vs Fatalities (Bulgaria)

**H0:** There is no statistically significant relationship between exposure (population as proxy) and road fatalities.  
**H1:** There is a statistically significant relationship between exposure and road fatalities.

To test this hypothesis, the Pearson correlation coefficient is applied. It should be noted that Pearson correlation measures association, not causation, and assumes a linear relationship between variables.

A significance level of α = 0.05 is used. The null hypothesis is rejected if the p-value is less than 0.05, indicating a statistically significant relationship between the variables.

The chosen significance level (α = 0.05) reflects a standard balance between Type I and Type II error risk commonly used in statistical analysis.

In cases where the p-value exceeds the chosen significance level (α = 0.05), we fail to reject the null hypothesis.

This does not imply that the null hypothesis is proven true, but rather that there is insufficient statistical evidence to support the alternative hypothesis.

This conservative interpretation is adopted to minimize the risk of Type I error (false positive), ensuring that any detected relationship is statistically robust rather than a result of random variation in the data.



```python
# recreate df_bg properly

fatalities_bg = Road_fatalities_BG[["Year", "fatalities"]].copy()
fatalities_bg = fatalities_bg.rename(columns={"Year": "year"})

population_bg = Population_bg_2009_2024[["Year", "Total"]].copy()
population_bg = population_bg.rename(columns={
    "Year": "year",
    "Total": "population"
})

df_bg = fatalities_bg.merge(population_bg, on="year", how="left")

df_bg["road_km"] = 19986

df_bg["fatalities_per_million"] = (
    df_bg["fatalities"] / df_bg["population"] * 1_000_000
)
```

```python
from scipy.stats import pearsonr

df_corr = df_bg.dropna(subset=["population", "fatalities"])

corr, p_value = pearsonr(
    df_corr["population"],
    df_corr["fatalities_per_million"]
)

print("Correlation:", corr)
print("p-value:", p_value)
```

```python
if p_value < 0.05:
    print("Statistically significant relationship detected")
else:
    print("No statistically significant relationship detected")
```

The Pearson correlation coefficient between the exposure proxy (population) and road fatalities is r = 0.585, indicating a moderate positive relationship.

The associated p-value (p = 0.017) is below the significance level of α = 0.05. Therefore, the null hypothesis is rejected, and it is concluded that there is a statistically significant relationship between the exposure proxy and road fatalities.

This suggests that higher levels of exposure, as approximated by population, are associated with increased numbers of fatalities.

However, this relationship should not be interpreted as causal. The exposure variable is represented by a proxy (population), which does not directly measure traffic intensity. Therefore, the results indicate statistical association rather than causation.

#### 12.2 Hypothesis 2: Bulgaria vs EU Benchmark Comparison

**H0:** There is no statistically significant difference in fatality rates between Bulgaria and the EU average.

**H1:** Bulgaria has a statistically significantly different (higher) fatality rate compared to the EU average.

To test this hypothesis, an independent samples t-test is applied, comparing the distribution of Bulgaria’s annual values against the corresponding EU benchmark values over the same time period.

A significance level of α = 0.05 is used. The null hypothesis is rejected if the p-value is below this threshold.

#### 12.2.1. Construct comparable dataset

```python
# Select only comparable entities
df_comp = df_eu.copy()

# Keep only Bulgaria and EU benchmark
df_comp = df_comp[df_comp["country"].isin(["Bulgaria", "EU"])].copy()

# Ensure correct types
df_comp["year"] = df_comp["year"].astype(int)
df_comp["fatalities_per_million"] = pd.to_numeric(df_comp["fatalities_per_million"], errors="coerce")

# Drop missing values
df_comp = df_comp.dropna()

df_comp.head()
```

#### 12.2.2. Statistical Test (Welch t-test)

```python
bg = df_comp[df_comp["country"] == "Bulgaria"]["fatalities_per_million"]
eu = df_comp[df_comp["country"] == "EU"]["fatalities_per_million"]

# Welch t-test (does NOT assume equal variance)
t_stat, p_value = stats.ttest_ind(bg, eu, equal_var=False)

print("t-statistic:", t_stat)
print("p-value:", p_value)

alpha = 0.05
if p_value < alpha:
    print("Statistically significant difference detected between Bulgaria and EU.")
else:
    print("No statistically significant difference detected between Bulgaria and EU at α = 0.05.")
```

#### 12.2.3. Effect Size (Cohen’s d)

```python
def cohens_d(x, y):
    nx, ny = len(x), len(y)
    pooled_std = np.sqrt(((nx - 1)*np.var(x, ddof=1) + (ny - 1)*np.var(y, ddof=1)) / (nx + ny - 2))
    return (np.mean(x) - np.mean(y)) / pooled_std

d = cohens_d(bg, eu)

print("Cohen's d:", d)
```

Effect Size Interpretation (Cohen’s d)

- d ≈ 0.2 → small difference
- d ≈ 0.5 → medium difference
- d ≥ 0.8 → large difference

Effect size provides a scale-independent measure of the difference between Bulgaria and the EU benchmark, complementing the p-value.

#### 12.2.5. Confidence Interval

```python
diff = np.mean(bg) - np.mean(eu)

se = np.sqrt(np.var(bg, ddof=1)/len(bg) + np.var(eu, ddof=1)/len(eu))

ci_low = diff - 1.96 * se
ci_high = diff + 1.96 * se

print("Mean difference (BG - EU):", diff)
print("95% CI:", (ci_low, ci_high))
```

```python
# Mean difference

diff = np.mean(bg) - np.mean(eu)

# Standard Error

se = np.sqrt(
    np.var(bg, ddof=1) / len(bg) +
    np.var(eu, ddof=1) / len(eu)
)


# Confidence Interval using ppf

confidence = 0.95

lower_area = (1 - confidence) / 2
upper_area = (1 + confidence) / 2

ci_low = st.norm.ppf(lower_area, loc=diff, scale=se)
ci_high = st.norm.ppf(upper_area, loc=diff, scale=se)


# Output

print(f"Mean difference (BG - EU): {diff:.2f}")
print(f"95% Confidence Interval: [{ci_low:.2f}, {ci_high:.2f}]")
```

Confidence Interval Interpretation
A 95% confidence interval for the difference in fatalities per million inhabitants between Bulgaria and the EU benchmark is constructed using the normal distribution quantile function (percent point function, ppf).

The mean difference (BG − EU) is used as the center of the distribution, while the standard error of the difference is used as the scale parameter.

The confidence interval is computed as:

$$ CI = ppf(\frac{\alpha}{2}, loc = diff, scale = SE), ppf(1 - \frac{\alpha}{2}, loc = diff, scale = SE) $$

where α = 0.05 for a 95% confidence level.

This interval provides a range of plausible values for the true difference between Bulgaria and the EU. If the interval includes zero, the difference is not statistically significant at the 5% level. If the interval does not include zero, the difference is considered statistically significant.

The normal approximation is used for simplicity, assuming that the sampling distribution of the mean difference is approximately normal.

#### 15.2.7. Confidence Interval with t-distribution

```python
diff = np.mean(bg) - np.mean(eu)


# Standard Error

se = np.sqrt(
    np.var(bg, ddof=1) / len(bg) +
    np.var(eu, ddof=1) / len(eu)
)


# Degrees of freedom (Welch-Satterthwaite)

n1, n2 = len(bg), len(eu)
s1, s2 = np.var(bg, ddof=1), np.var(eu, ddof=1)

df = (s1/n1 + s2/n2)**2 / (
    (s1/n1)**2 / (n1 - 1) +
    (s2/n2)**2 / (n2 - 1)
)


# t critical value

confidence = 0.95
alpha = 1 - confidence

t_crit = st.t.ppf(1 - alpha/2, df)


# Confidence Interval

ci_low = diff - t_crit * se
ci_high = diff + t_crit * se


# Output

print(f"Mean difference (BG - EU): {diff:.2f}")
print(f"95% CI (t-distribution): [{ci_low:.2f}, {ci_high:.2f}]")
print(f"Degrees of freedom: {df:.2f}")
```

A 95% confidence interval for the difference in fatalities per million inhabitants between Bulgaria and the EU benchmark is constructed using the Student's t-distribution.

The mean difference (BG - EU) is used as the center of the interval, and the standard error of the difference is used as the scale parameter.

The critical value is obtained from the t-distribution using the Welch-Satterthwaite approximation for the degrees of freedom, which accounts for unequal variances between the two groups.

The confidence interval is computed as:

$$ CI = diff ± t_{critical} x SE $$

This approach is more appropriate than the normal approximation due to the relatively small sample size and potential differences in variance between the compared groups.

If the confidence interval includes zero, the difference is not statistically significant at the 5% level. If it excludes zero, the difference is considered statistically significant.
